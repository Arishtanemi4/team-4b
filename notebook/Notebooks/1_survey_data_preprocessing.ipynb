```python
import pandas as pd
from pathlib import Path


def parse_rater(rater_id) -> tuple[int, int]:
    """
    Decode a Rater ID of form ±TTPPPPP into (team, person).

    Examples
    --------
    >>> parse_rater(-100001)   # team 1, person 1
    (1, 1)
    >>> parse_rater(-2500005)  # team 25, person 5
    (25, 5)
    """
    if pd.isna(rater_id):
        return (None, None)
    a = abs(int(rater_id))
    team   = a // 100_000
    person = a % 10
    return team, person


def load_with_rater(path: str | Path, sheet: str = "By Rater") -> pd.DataFrame:
    """Load a snapshot file (csv or xlsx) and add team + person columns."""
    path = Path(path)
    if path.suffix == ".csv":
        df = pd.read_csv(path, header=1)
    else:
        df = pd.read_excel(path, sheet_name=sheet, header=1)

    # First column is Rater but usually reads as 'Unnamed: 0'
    df = df.rename(columns={df.columns[0]: "Rater"})
    df = df[pd.notna(df["Rater"])].copy()
    df["Rater"] = df["Rater"].astype(int)

    decoded = df["Rater"].apply(parse_rater)
    df["team"]   = [t for t, _ in decoded]
    df["person"] = [p for _, p in decoded]
    return df
```
